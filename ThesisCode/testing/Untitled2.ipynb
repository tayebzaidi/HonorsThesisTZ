```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 150, 150)
under = np.sin(x/10)
```

```python
plt.plot(under)
plt.show()
```

```python
x2 = np.linspace(0,57, 150)
under2 = np.sin(x2/5)
plt.plot(x2, under2)
plt.show()
```

```python
ind1 = np.arange(len(under))
ind2 = np.arange(len(under2))
print(ind1)
print(ind2)
plt.plot(ind1, under)
plt.plot(ind2, under2)
plt.show()
```
